#Z-score

```python
# prompt: calculate Z-score for df2 X2

import numpy as np

# Assuming df2 is defined as in the provided code:
data = [[1, 200], [-2, 400], [0, 300], [-3, -200]]
df2 = pd.DataFrame(data, columns=['X1', 'X2'])

# Calculate the Z-score for 'X2'
df2['X2_zscore'] = (df2['X2'] - df2['X2'].mean()) / df2['X2'].std()

df2
```
